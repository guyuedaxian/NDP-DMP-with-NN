```python
def read_file(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
    line = lines[0].replace('[','').replace(']','').replace('"', '').split(',')
    line_values = []
    for i in range(0, len(line)):
        line_values.append(float(line[i]))
    
    return line_values
```

### 1, Mean Loss Values of Models

```python
def mean_loss(label, paths):
    file_name   = 'test_loss.txt'
    value = 0
    for p in paths:
        path = root_path + '/ndp/ndp_models/' +p  + file_name
        y = read_file(path)
        value = value + y[-1]
        print(label, y[-1], value)
        
    print(label, round(value/len(paths), 4))
        
```
